```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# style visual
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (9,5)

# load dataset
tips = sns.load_dataset("tips")

# tampilkan data
tips.head()
```

```python
plt.figure(figsize=(9,5))

ax = sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    estimator='mean',
    palette="viridis"
)

# menampilkan angka di atas bar
for i in ax.containers:
    ax.bar_label(i, fmt='%.1f')

plt.title("Average Total Bill per Day", fontsize=14, weight='bold')
plt.xlabel("Day")
plt.ylabel("Average Total Bill")

plt.show()
```

```python
plt.figure(figsize=(9,5))

sns.histplot(
    data=tips,
    x="total_bill",
    hue="time",
    bins=20,
    kde=True,
    color="skyblue"
)

plt.title("Distribution of Total Bill", fontsize=14, weight='bold')
plt.xlabel("Total Bill")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(9,5))

sns.boxplot(
    data=tips,
    x="smoker",
    y="tip",
    palette="Set2"
)

plt.title("Tip Distribution by Smoker Status", fontsize=14, weight='bold')
plt.xlabel("Smoker Status")
plt.ylabel("Tip")

plt.show()
```

```python
plt.figure(figsize=(9,5))

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time",
    palette="coolwarm",
    s=100
)

# Add regression line
sns.regplot(
    data=tips,
    x="total_bill",
    y="tip",
    scatter=False,
    color='black'
)

plt.title("Relationship Between Total Bill and Tip", fontsize=14, weight='bold')
plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```
