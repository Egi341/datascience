```python
import pandas as pd
import numpy as np
```

```python
# memuat data titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df  = pd.read_csv(URL)
```

```python
# explorasi awal
display(df.head())
display(df.shape)
```

```python
# jumlah baris dan kolom
rows, cols = df.shape
print(f"Jumlah baris: {rows}")
print(f"Jumlah kolom: {cols}")
```

```python
# daftar nama kolom
print("Daftar nama kolom:")
for col in df.columns:
    print(f"- {col}")
```

```python
# missing vslue
missing = df.isna().sum()
missing = missing[missing > 0]
max_len = max(len(col) for col in missing.index)

print("Kolom dengan missing values:")
for col, val in missing.items():
    print(f"- {col.ljust(max_len)} : {val}")
```

```python
# filter penumpang wanita kelas 1
female_class1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
survived = female_class1[female_class1['Survived'] == 1]
survived_count = survived['Survived'].count()
avg_age = survived['Age'].mean()

print("Penumpang wanita kelas 1 yang selamat")
print(f"- Jumlah yang selamat : {survived_count} orang")
print(f"- Rata-rata usia      : {avg_age:.2f} tahun")
```

```python
# kelas dengan keselamatan tertinggi
survival_rate = df.groupby('Pclass')['Survived'].mean()

print("Tingkat keselamatan per kelas:")
for pclass, rate in survival_rate.items():
    print(f"- Kelas {pclass} : {rate:.2%}")
best_class = survival_rate.idxmax()
best_rate = survival_rate.max()
print(f"Kelas dengan tingkat keselamatan tertinggi adalah Kelas {best_class} ({best_rate:.2%})")
```

```python
# Q1 Berapa jumlah baris dan kolom dataset Titanic?
rows, cols = df.shape

print("Ukuran dataset Titanic:")
print(f"- Jumlah baris : {rows}")
print(f"- Jumlah kolom : {cols}")
```

```python
# Q2 Kolom mana yang memiliki nilai kosong (missing)?
missing_cols = df.columns[df.isna().any()]

print("Kolom dengan missing values:")
for col in missing_cols:
    print(f"- {col}")
```

```python
# Q3 Berapa persen penumpang yang selamat (Survived=1)?
survival_rate = df['Survived'].mean() * 100

print(f"Persentase penumpang yang selamat: {survival_rate:.2f}%")
```

```python
# Q4 Rata-rata usia penumpang wanita kelas 1 yang selamat?
avg_age = df[
    (df['Sex'] == 'female') &
    (df['Pclass'] == 1) &
    (df['Survived'] == 1)
]['Age'].mean()

print(f"Rata-rata usia penumpang wanita kelas 1 yang selamat: {avg_age:.2f} tahun")
```

```python
# Q5 Kelas penumpang mana yang tingkat  keselamatan tertinggi?
best_class = df.groupby('Pclass')['Survived'].mean().idxmax()

print(f"Kelas dengan tingkat keselamatan tertinggi: Kelas {best_class}")
```
