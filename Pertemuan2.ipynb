```python
import pandas as pd
import numpy as np

URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(URL)
df.head()
```

Jumlah baris dalam kolom pada dataset

```python

baris, kolom = df.shape
print(f"Jumlah baris: {baris}")
print(f"Jumlah kolom: {kolom}")

for index, column_name in enumerate(df.columns):
    print(f"{index+1}. {column_name}")
```

kolom yang memiliki nilai kosong

```python
print(df.isnull().sum())
```

Persentase penumpang selamat

```python
survived = df['Survived'].mean() * 100
print(f"Percentage penumpang yang selamat adalah {survived:.2f}%")
```

Rata-rata usia penumpang wanita keas 1 yang selamat

```python
filtered_passanger = df[(df['Sex'] == 'female') & (df['Pclass'] == 1) & (df['Survived'] == 1)]
average_age = filtered_passanger['Age'].mean()
print(f"Rata-rata penumpang wanita kelas 1 yang selamat adalah {average_age:.2f}")
```

Penumpang dengan tingkat keselamatan tertinggi

```python
most_survived = df.groupby('Pclass')['Survived'].mean().idxmax()
print(f"Kelas penumpang dengan tingkat keselamatan tertinggi adalah {most_survived}")
```

tingkat keselamatan perkelas

```python
rate_survived = df.groupby('Pclass').agg(total = ('PassengerId', 'count'), survived = ('Survived', 'mean')).round(3)
print(rate_survived)
```
